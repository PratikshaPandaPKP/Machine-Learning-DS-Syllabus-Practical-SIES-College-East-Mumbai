```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import Isomap
from sklearn.datasets import make_swiss_roll

# Generate Swiss Roll dataset (3D)
n_samples = 1000
X, color = make_swiss_roll(n_samples, noise=0.1)

# Apply Isomap to reduce 3D to 2D
isomap = Isomap(n_neighbors=10, n_components=2)
X_transformed = isomap.fit_transform(X)

# Plot Original 3D Data
fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap='Spectral')
ax1.set_title("Original 3D Data (Swiss Roll)")

# Plot 2D Transformed Data
ax2 = fig.add_subplot(122)
ax2.scatter(X_transformed[:, 0], X_transformed[:, 1], c=color, cmap='Spectral')
ax2.set_title("Isomap Reduced 2D Data")

plt.show()
```
